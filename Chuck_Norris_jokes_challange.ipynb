```python
##chuck noris jokes api
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup as soup
url = "http://api.icndb.com/jokes?"
parameters = {
    'id' : '15'
}
response = requests.get(url)
r = response.content
#rint(type(r))
#rint(response.content.decode('utf-8'))
json_Data = json.loads(r)

        
        
json_value = json_Data['value']


#print(json_value)

id_list = []
joke_list = []
for i in range(474):
     id_list.append(json_value[i]['id'])



for i in range(474):
     joke_list.append(json_value[i]['joke'])
    


chuck_dict = dict(zip(id_list,joke_list))
#print(chuck_dict)


current_list = []
main_list = []
with open('id.csv','r') as fp:
    id_csv = fp.read()[2:]          #reading from test cases and conveting it to different list by comparing value
    for id_csv in chuck_dict:
        current_list.append(id_csv)    #list of ids
        main_list.append(chuck_dict[id_csv])  #list of joke
 

#print(current_list)
#print(main_list)
lost_list = []
for i in main_list:
    lost_list.append( i.replace(",",""))
    
    

data = {
    'id':current_list,
    'joke':main_list       
    
}
 
df = pd.DataFrame(data , columns=['id','joke'])   #creating a dataframe

df.to_csv('chuck.csv',index=False) #conveting it to csv file using pandas
    

    




















```
